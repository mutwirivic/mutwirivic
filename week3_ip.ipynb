```
#Importing required libraries
import pandas as pd
import numpy as np
```

```
#Loading the required datasets
#cells geo dataset
cells_geo_df = pd.read_csv('cells_geo.csv', delimiter=';',index_col=0)
cells_geo_df
```

```
#Cells geo decription
cells_geo_description_df = pd.read_excel('cells_geo_description.xlsx',sheet_name = 0)
cells_geo_description_df
```

```
#CDR description
cdr_df = pd.read_excel('CDR_description.xlsx', sheet_name= 0)
cdr_df.head()
```

```
#Telcom 1
t1_df = pd.read_csv('Telcom_dataset.csv')
t1_df.columns = ['PRODUCT','VALUE','DATETIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
t1_df.head()
```

```
#Telcom2
t2_df = pd.read_csv('Telcom_dataset2.csv')
t2_df.columns = ['PRODUCT','VALUE','DATETIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
t2_df.head()
```

```
#Telcom3
t3_df = pd.read_csv('Telcom_dataset3.csv')
t3_df.columns = ['PRODUCT','VALUE','DATETIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
t3_df.head()
```

```
#Merge all telcom dataframes into one dataframe
telcom_df = pd.concat([t1_df,t2_df,t3_df])
telcom_df
```
